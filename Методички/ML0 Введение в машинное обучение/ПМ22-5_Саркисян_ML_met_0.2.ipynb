Найдите процентное соотношение учащихся, сдавших экзамен на «отлично», «хорошо», «удовлетворительно», «неудовлетворительно».

```python
#7
def replace_func(x):
  if x < 63:
    return 2
  elif x > 62 and x < 77:
    return 3
  elif x > 63 and x < 91:
    return 4
  else:
    return 5
```

```python
df['Балл 5'] = df.loc[:, 'Балл']
df
```

```python
df['Балл 5'] = df['Балл 5'].apply(replace_func)
df
```

```python
df.groupby(['Балл 5']).agg({'Балл 5' : 'count'}) / df.shape[0] * 100
```

```python
#11
for i in range(short_ans):
  df.insert(len(df.columns), f'B{i+1}', 0)

df
```

```python
for i, _ in df.iterrows():
  short_ans = list(df.loc[i, 'Задания с кратким ответом'])
  for j in range(len(short_ans)):
    df.loc[i, f'B{j+1}'] = short_ans[j]

df
```
