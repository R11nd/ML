1. С помощью массивов numpy создайте таблицу умножения.

```python
import numpy as np

n = 9

table = np.empty((n, n), dtype=int)

for i in range(n):
    for j in range(n):
        table[i, j] = (i + 1) * (j + 1)

print(table)
```

2. Создайте функцию, которая принимает как аргументы целое число N и первый элемент (вещественное число), и разность (вещественное число) и создает матрицу numpy по диагонали, которой располагаются первые N членов арифметической прогрессии.

```python
def create_ap_matrix(N, first_element, difference):
    progression = np.arange(first_element, first_element + N * difference, difference)

    matrix = np.zeros((N, N), dtype=float)

    np.fill_diagonal(matrix, progression)

    return matrix

N = 5  # Количество членов прогрессии
first_element = 2.0  # Первый элемент прогрессии
difference = 3.0  # Разность между членами прогрессии

result_matrix = create_ap_matrix(N, first_element, difference)
print(result_matrix)
```

5. Создайте две матрицы размером (5,5). Одна матрица содержит 5 в шахматном порядке как в задаче домашнего задания, другая имеет треугольную форму содержащую 5 на основной диагонали и в позициях выше ее, а ниже все 0. Посчитайте их детерминант и найдите обратные матрицы.

```python
matrix_chess = np.array([[5, 0, 5, 0, 5],
                         [0, 5, 0, 5, 0],
                         [5, 0, 5, 0, 5],
                         [0, 5, 0, 5, 0],
                         [5, 0, 5, 0, 5]])

# создание треугольной матрицы с 5 на основной диагонали и нулями выше и ниже
matrix_triangle = np.tri(5, 5) * 5

# находим детерминанты
det_chess = np.linalg.det(matrix_chess)
det_triangle = np.linalg.det(matrix_triangle)

# находим обратные матрицы
try:
    inv_chess = np.linalg.inv(matrix_chess)
    inv_triangle = np.linalg.inv(matrix_triangle)
except np.linalg.LinAlgError:
    inv_chess = "Обратной матрицы не существует"
    inv_triangle = "Обратной матрицы не существует"

# вывод
print("Матрица с 'шахматным' распределением числа 5:")
print(matrix_chess)
print("Детерминант матрицы с 'шахматным' распределением числа 5:", det_chess)
print("Обратная матрица с 'шахматным' распределением числа 5:")
print(inv_chess)

print("\nТреугольная матрица с 5 на основной диагонали:")
print(matrix_triangle)
print("Детерминант треугольной матрицы:", det_triangle)
print("Обратная треугольная матрица:")
print(inv_triangle)
```
